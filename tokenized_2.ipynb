```python
import pandas as pd
```

```python
data = pd.read_pickle("nneew")
```

```python
data.head()
```

```python
original_word_df = data.filter(['new_title_word','javascript_code','javascript_tokenized', 'title_word_tokenized'], axis=1)
```

```python
original_tag_df = data.filter(['new_title_tag','javascript_code','javascript_tokenized', 'title_tag_tokenized'], axis=1)
```

```python
original_title_df = data.filter(['title','javascript_code','javascript_tokenized', 'title_tokenized'], axis=1)
```

```python
def modify_df(df):
    df["repo"] = ""
    df["path"] = ""
    df["func_name"] = ""
    df["url"] = ""
    df["sha"] = ""
    df["partition"] = ""
    # new_df["original_string"] = ""
    df["language"] = "javascript"
    df["original_string"] = df["javascript_code"]
    return df
```

```python
# original_word_df["repo"] = ""
# original_word_df["path"] = ""
# original_word_df["func_name"] = ""
# original_word_df["url"] = ""
# original_word_df["sha"] = ""
# original_word_df["partition"] = ""
# # new_df["original_string"] = ""
# original_word_df["language"] = "javascript"
# original_word_df["original_string"] = original_word_df["javascript_code"]
```

```python
original_word_df = modify_df(original_word_df)
original_tag_df = modify_df(original_tag_df)
original_title_df = modify_df(original_title_df)
```

```python
original_word_df.rename(columns={'javascript_code':'code'}, inplace=True)
original_word_df.rename(columns={'javascript_tokenized':'code_tokens'}, inplace=True)
original_word_df.rename(columns={'new_title_word':'docstring'}, inplace=True)
original_word_df.rename(columns={'title_word_tokenized':'docstring_tokens'}, inplace=True)
```

```python
original_tag_df.rename(columns={'javascript_code':'code'}, inplace=True)
original_tag_df.rename(columns={'javascript_tokenized':'code_tokens'}, inplace=True)
original_tag_df.rename(columns={'new_title_tag':'docstring'}, inplace=True)
original_tag_df.rename(columns={'title_tag_tokenized':'docstring_tokens'}, inplace=True)
```

```python
original_title_df.rename(columns={'javascript_code':'code'}, inplace=True)
original_title_df.rename(columns={'javascript_tokenized':'code_tokens'}, inplace=True)
original_title_df.rename(columns={'title':'docstring'}, inplace=True)
original_title_df.rename(columns={'title_tokenized':'docstring_tokens'}, inplace=True)
```

```python
original_word_df.to_pickle("word_df_dif_tokens")
```

```python
original_tag_df.to_pickle("tag_df_dif_tokens")
```
